```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
```

```python
gold_data = pd.read_csv('gld_price_data.csv')
```

```python
gold_data.head() #after doing a randomforestclass, try forecasting models 
```

```python
gold_data.isna().sum()
```

```python
gold_data.describe()
```

```python
#creating correlation matrix to see if any correlations exist between columns
correlation_matrix = gold_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.scatter(gold_data['SLV'], gold_data['GLD'], label='Actual')
plt.xlabel('SLV')
plt.ylabel('GLD')
```

```python
sns.distplot(gold_data['GLD'], color = 'blue')
```

##### Split feautures and labels

```python
X = gold_data.drop(columns = ['Date', 'GLD'], axis = 1)
Y = gold_data['GLD']
```

```python
print(X,Y)
```

##### Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .1)
```

##### Model Training

```python
model = RandomForestRegressor(n_estimators = 1000)
```

```python
model.fit(X_train, Y_train)
```

##### Model Evaluation

```python
X_test_prediction = model.predict(X_test)
```

```python
#r squared error and MAE
error_score = metrics.r2_score(Y_test,X_test_prediction) #r squared is closer to 1 signifying a really accurate prediction
print('R squared error:', error_score)
mae = mean_absolute_error(Y_test, X_test_prediction)
print('Mean Absolute error:', mae) #mae is closer to 0 meaning its really accurate
```

##### Compare actual vs predicted values in a plot

```python
Y_test = list(Y_test)
plt.plot(Y_test, color = 'blue', label = 'Actual Value')
plt.plot(X_test_prediction, color = 'green', label = 'Predicted Value')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
```
