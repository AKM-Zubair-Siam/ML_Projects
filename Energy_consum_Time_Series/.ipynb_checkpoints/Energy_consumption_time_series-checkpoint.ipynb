```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('PJME_hourly.csv')
data = data.set_index('Datetime')
data.index = pd.to_datetime(data.index)
```

```python
data
```

```python
data.plot(style = '.', title = 'PJME Energy Use')
```

###### Train Test Split

```python
train = data.loc[data.index < '01-01-2015']
test = data.loc[data.index >= '01-01-2015']

fig, ax = plt.subplots(figsize = (15,5))
train.plot(ax=ax, label = 'Train')
test.plot(ax = ax, label = 'Test')
```

```python
data.loc[(data.index > '01-01-2010') & (data.index < '01-08-2010')].plot()
```

###### Feature Creation

```python
def create_features(data):
#creating time series index based on dataframe
    data = data.copy()
    data['hour'] = data.index.hour
    data['day_of_week'] = data.index.day_of_week
    data['quarter'] = data.index.quarter
    data['month'] = data.index.month
    data['year'] = data.index.year
    data['day_of_year'] = data.index.day_of_year
    return data

data = create_features(data)
```

```python
#Visualizing features
```

```python
sns.boxplot(data = data, x = 'hour', y = 'PJME_MW')
```

```python
sns.boxplot(data = data, x = 'month', y = 'PJME_MW')
```

###### Model Creation

```python
train = create_features(train)
test = create_features(test)
```

```python
data.columns
```

```python
X = ['hour', 'day_of_week', 'quarter', 'month', 'year', 'day_of_year' ]
Y = 'PJME_MW'
```

```python
X_train = train[X]
Y_train = train[Y]
X_test = test[X]
Y_test = test[Y]
```

```python
reg = xgb.XGBRegressor(n_estimators = 1000, early_stopping_rounds = 50, learning_rate = .01)
reg.fit(X_train, Y_train, 
        eval_set = [(X_train, Y_train), (X_test, Y_test)],
        verbose = 100
       )
```

###### Feature Importance

```python
fi = pd.DataFrame(reg.feature_importances_,
             index = reg.feature_names_in_,
             columns = ['importance'])
```

```python
fi.sort_values('importance').plot(kind = 'barh', title = 'Feature Importance')
```

###### Forecast on Test

```python
test['prediction'] = reg.predict(X_test)
```

```python
data.merge(test['prediction'], how = 'left', left_index = True, right_index = True)
```

```python
score = np.sqrt(mean_squared_error(test['PJME_MW'], test['prediction']))
```

```python
score
```

```python
test['error'] = np.abs(test[Y] - test['prediction'])
```

```python
test['date'] = test.index.date
```
