```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset = pd.read_csv('boston.csv')
```

```python
house_price_dataset.head()
```

```python
house_price_dataset.rename(columns={'MEDV': 'price'}, inplace=True)
```

```python
house_price_dataset.head(1)
```

```python
house_price_dataset.shape
```

```python
house_price_dataset.isnull().sum() #checks for null values in all columns
```

```python
house_price_dataset.describe()
```

```python
#creating correlation matrix to see if any correlations exist between columns
correlation_matrix = house_price_dataset.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

Splitting the data and target

```python
X = house_price_dataset.drop(columns = 'price', axis = 1)
Y = house_price_dataset['price']
```

```python
print (X,Y)
```

Split data into training and test data

```python
#IMPORTANT NOTE: youre not using stratify = y here because y label has multiple values and not binary values. 
#so proportionally splitting them doesnt make sense here
#if y label had values of 0,1 or "spam" or "not spam", then you need to use stratify to ensure proportinal class distribution
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .3) 
print(X.shape, X_train.shape, X_test.shape)
```

Model Training time


XGBoost Regressor Model

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
#prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R Squared error for training data
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute error for training data
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print('R Squared is:', score_1)
print('Mean Absolute error is:', score_2)
```

Visualize actual prices vs predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Price: training data")
plt.show()
```

```python
test_data_prediction = model.predict(X_test)
```

```python
# R Squared error for test data
score_3 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute error for test data
score_4 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print('R Squared is:', score_3)
print('Mean Absolute error is:', score_4)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Price: test data")
plt.show()
```
